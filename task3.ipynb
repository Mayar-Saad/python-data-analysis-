```python
vowel=["a","e","i","o","u","A","E","I","O","U"]
```

```python
def character_is_avowel_or_consonant(char):
    if char in vowel:
        return"char is vowel"
    elif type(char)==str:        
        return"char is constant"
    else:
        return"wrong value"
```

```python
#•	Check if a password is strong (length > 8)
```

```python
def is_the_password_is_strong(no):
    no = str(no)
    if len(no)>=8:
        return "strong"
    else:
        return"weak"
```

```python
#•	Determine the grade based on marks
```

```python
def your_grade(marks):
    if marks >=85:
        return"your grade is A"
    elif marks>=70:
        return"your grade is B"
    elif marks>=65:
        return"your grade is C"
    elif marks>=50:
        return"your grade is D"
    else:
        return"Failed"
```

```python
your_grade(75)
```
